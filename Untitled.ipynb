```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
#Load csv file
df = pd.read_csv('heart_disease_data.csv')
df
```

```python
df.head()
```

```python
# Calculate the average of a selected column 
average_age = df['age'].mean()
print(f"Average Age: {average_age}")
```

```python
# Check for missing values in the dataframe
missing_values = df.isnull().sum()
print(f"Missing Values:\n{missing_values}")
```

```python
# Get general statistics of the numerical columns
statistics = df.describe()
print(f"Statistics:\n{statistics}")
```

```python
# Bar Chart for the 'cp' column
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 4))
df['cp'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Chest Pain Type Distribution')
plt.xlabel('Chest Pain Type')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python
# Scatter Plot for 'age' vs 'thalach'
plt.figure(figsize=(8, 4))
plt.scatter(df['age'], df['thalach'], alpha=0.5, color='purple')
plt.title('Age vs Maximum Heart Rate Achieved')
plt.xlabel('Age')
plt.ylabel('Maximum Heart Rate Achieved')
plt.show()
```

```python
# Heatmap for correlation matrix
import seaborn as sns

plt.figure(figsize=(10, 6))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=1)
plt.title('Correlation Matrix')
plt.show()
```
